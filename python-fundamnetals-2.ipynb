# Implicit Conversion
- Implicit data type conversion is done automatically by the python compiler without any involvement of the user.

```python
Revenue_1 = 12
Revenue_2 = 15.5

# knowing the variable types
print("The variable type is", type(Revenue_1))
print("the variable type is", type(Revenue_2))

# knowing the total Revenue
print("The total revenue is", Revenue_1 + Revenue_2)

# storing the outcome in a variable
Revenue =  Revenue_1 + Revenue_2 # Here it is converting to float

print("The total revenue is", Revenue) # Here it is converting to string
print("The variable type for revenue is",type( Revenue))

# Here in the above cases, implicit conversion is taking place
```

# Explict Conversion

- The conversion of one data type into another, done via user intervention or manually as per the requirement.


```python
sales_A = input("Enter store A sales")
sales_B = input("Enter store B sales")
print("The type of input is",type(sales_A))
print("The type of input is",type(sales_B))

Total_sales = sales_A + sales_B
print("The type of output it is generating is",type(Total_sales))
print("The total sales generates is", Total_sales)


# Here joining / concatenation is taking place
```

```python
Total_sales1 = float(sales_A) + float(sales_B) #Explict conversion

print("Total sales generated is",Total_sales1,", Here the type of variable is",type(Total_sales1))


# In the below situation we might lose the precison in case of decimals.
Total_sales2 = int(sales_A) + int(sales_B) #Explict conversion


print("Total sales generated is",Total_sales2,", Here the type of variable is",type(Total_sales2))


```

# Arithmetic Operators

- Addition
- Subraction
- Mutiplication
- Division
- Modulus Division
- Floor Division
- Exponentiation

```python
#addition
sale1 = int(input("The sales of shop 1"))
sale2 = int(input("The sales of shop 2"))
sales = sale1 + sale2 

#subraction
diff_sales = sale1 - sale2

print("The total sales", sales, "The total difference in sales", diff_sales)
```

```python
#multiplication
sales_per_unit = int(input("Enter the sales per unit price"))
no_of_units = int(input("Enter the number of units"))
Total = sales_per_unit * no_of_units

print("Total sales generated on sold",no_of_units," units is",Total)
```

```python
#division
yearly_sales = int(input("The yearly sales"))
avg_sales_per_month = yearly_sales / 12 

print("The average sales of a month",avg_sales_per_month)
```

```python
#modulus_division - It is used to find reminder
v = int(input("Enter a number to know if it is Even or Odd ?"))
reminder = v % 2

if reminder == 1:
    print("odd")
else:
    print("even")
```

```python
#floor_divison - It is used to find quotient    
students = int(input("Enter total number of students"))
batches = int(input("Enter desired number of batches to have"))

quotient = students // batches
left = students % batches

if left == 0:
    print("The total number of batches can be formed is", quotient)
else:
    print("The total number of batches can be formed is", quotient, "and the remaning students will be", left)
```

```python
#exponentitaion
a = int(input("Enter the number you would like to perform exponential"))
b = int(input("Enter the exponential power you would like to know"))

c = a**b

print("The Value of", a , "exponentital", b, "is", c)
```
