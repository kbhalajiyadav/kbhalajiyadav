**Q. Print prime numbers for a given range**

```python
n1 = int(input("Enter the lower limit number"))
n2 = int(input("Enter the upper limit number"))

if n1 == 1:
    n1 = n1 +1
    
i = n1
while i <= n2:                 # range of number which we have to find weather they are prime or not
    j = n1
    while j <= i // 2 + 1:     # maximun multiple we will check
                               # here we gives the range to iterate (be i // 2 + 1) the loop 
        if i == 2:
            print(i, end = ", ")
        if i % j == 0:
            break              # here it passes over to next if, if it is not satisfied
        if j == i // 2 + 1:    # here it checks if the it have runed i // 2 + 1 times, then print
            print(i, end = ", ")
        j = j + 1
    i = i + 1
```
