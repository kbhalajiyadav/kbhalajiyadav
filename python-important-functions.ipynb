# Comprehension
- It is an elegant way to create list, dictionary, sets in a concise manner.

```python
l1 = []

for i in range(1,6):
    l1.append(i**2)
print(l1)
```

```python
l2  = [i**2 for i in range(1,6)]   # This is how we can make it uch simpler
print(l2)
```

```python
l3 = []

for i in range(1,11):
    if i%2 ==0:
        l3.append(i)
        
print(l3)
```

# Using list comprehension

```python
l4  = [i for i in range(1,11) if i%2 == 0]
print(l4)
```

# Using Dictionary Comprehension

```python
d3 = {i:i**2 for i in range(1,6)}
print(d3)
```

```python
d4 = {
    "Aman":20,
     "Arul":40,
     "Ameer":22
     }
```

```python
d5 = {}

for i in d4:
    if d4[i] == 22 :
        d5[i] = "Yes"
    else:
        d5[i] = "No"
print(d5)
```

```python
d5  = {key: "Yes" if value > 22 else "No"
      for key,value in d4.items()}       # Here items will return the key values

print(d5)
```

# Lambda Function
    It also known as anonymus function
    
1. - Many arguments but one expression
- Take less time
- Used when function is needed for short duration
- often used with other functions

# sum 2 numbers

```python
def add(a,b):
    return a+b
    
a = add(2,3)
print(a)
```

```python
add1 = lambda p,q : p+q   # this is how we can perform it in simpler way

add1(2,3)
```

```python
add2 = lambda p,q,r,s : (p+q) / (r+s)  # this is how we can perform it in simpler way

s = add2(10,20,2,3)

print(s)
```

# Sort Function

```python
l1  = [2,3,5,5,6,2,1]
l1.sort(reverse = True) # Decending order

print(l1)
```

```python
l2 = [[1,2],[6,3],[2,0],[4,0]]
l2.sort()                       # Here it sorted based on first element
l2
```

```python
def s(a):
    return a[1] + a[0]
```

```python
l2 = [[1,2],[6,3],[2,0],[4,0]]
l2.sort(key = s)                       # Here it sorted based on sum of list elements
l2

# Keys can be used with functions, with the function in above cell where now it sorts with the second element in the list
```

```python
l2 = [[1,2],[6,3],[2,0],[4,0]]
l2.sort(key = lambda a: a[0] + a[1], reverse = True)                       # Here it sorted based on sum of elements in the list, in desending order
l2
```

# Map Function

```python
def attendance(attendance):
    if attendance > 80:
        return "Good"
    elif attendance > 60:
        return "OK"
    else:
        return "Need to Improve"
```

```python
attendance(30)  # calling an the function
```

# What if we have a list
    we can use a loop right?


```python
l1 = [30,40,50,60,90]

for i in l1:
    print(attendance(i))
```

# using map function

```python
l2 = [30,40,50,60,90]

# map (function iterable1, iterable2,.....)
print(map(attendance,l1))             # In map funtion we will place function first and then iterable

# Here it will work as the above for loop, where it will call the element in list and store it



print(set(map(attendance,l1)))   # here in sets duplicates will be removed

print(tuple(map(attendance,l1)))

print(list(map(attendance,l1)))

```

# Filter Function
-  It is used to filter a sequeance with the help of function. Only those values will be returned where answer is true.

```python
def check(a):
    vowels = ["a","e","i","o","u"]    # lits are iterable
    if a in vowels:
        return True
    else:
        return False
```

```python
check("l")
```

```python
l1 = ["a","b","e","i","q"]

m = list(map(check,l1))
print(m)
```

```python
f1 = list(filter(check,l1))  # here it has returned only the values in the list where it matches the vowels, it results in type of object so we convert it to a list

print(f1)
```

```python
l1 = [1,3,4,5,6]
list(filter(lambda u:u % 2, l1))
```
