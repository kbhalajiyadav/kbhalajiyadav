# Concentrate 2 strings index wise

```python
first_names = ["Modi","Amar","Roshan","Pankaj"]
last_names  = ["Amit","Sha","Raj","Sharma"]
l = len(first_names)

name = []

for i in range(l):
    x = first_names[i] + " " + last_names[i]
    name.append(x)
    
print(name)
```

# How many times a particular item is there in a list

```python
l = [2,2,2,1,1,23,12]
num = int(input("Enter the number"))
count = 0

for l in l:
    if num == l:
        count = count + 1
        
print(num, "is occuring", count,'times')
```

```python
l = [2,2,2,1,1,23,12]
num = int(input("Enter the number"))

print(num, "is occuring",l.count(num), "times")
```

# Check if we have a common items in 2 lists

```python
l =  [2,2,2,1,1,23,12]
l2 = [2,3,21,22,11, 41]

for i in l:
    for j in l2:
        if i == j:
            print(i)
            break
```

# Remove duplicates value from the list

```python
l2 = [2,3,21,22,11,41,2]

for i in l2:
    count = 0
    for j in l2:
        if i == j:
            count = count + 1
            if count > 1:
                l2.remove(j)
print(l2)
```
