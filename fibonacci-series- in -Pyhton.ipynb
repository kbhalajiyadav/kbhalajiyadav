The sequence of **Fibonacci numbers** can be defined as: Fn = Fn-1 + Fn-2

- The next number in the sequence, is the sum of the previous two numbers present in the sequence, starting from 0 and 1.

```python
n1 = 1
n2 = int(input("Enter the range ending number"))

a = 0 
b = 1

for i in range(n1,n2):
    print(a, end = ", ")   # "|n" is the default behaviour (to new line)
    c = b
    b = b + a
    a = c
```
