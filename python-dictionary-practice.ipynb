**Write a Python script to generate and print a dictionary that contains a number (between 1 and n) in the form (x,x^x)**

(1:1,2:4,3:9,4:16...)

```python
n =  int(input("enter the number of squares you would like to get")) 
D = {}

i = 1

for i in range(1,n+1): 
        D[i] = i *i
print(D)
```

**Write a Python program to map two lists into dictionary**

```python
name = ["Ameer", "Prathik", "Sanjay"]
age  = [26, 24, 28]

D2 = {}
j = 0

for i in name:
    D2[i] = age[j]
    j = j+1  
    
print(D2)  
```

```python
print(D2["Ameer"]) # fetching the data
```

**Write a program to sort the dictionary by keys**

```python
D3 = {
    23:"a",
    12:"b",
    76:"c",
    54:"d"
}

s = sorted(D3, reverse = True) # it is an generic function, for descending order (,reverse = True)
print(s)

D4 = {}

for i in s:
    D4[i] = D3[i]
    
print(D4)  # as we know here keys are used instead to index in dictionary
```

**Check how many times a word is theere in a senstence**

```python
a = "Fear leads to anger and anger leads to hate"
b = a.split()
D5 = {}

for i in b:
    if D5.get(i) == None: # to create fresh list in the list
        D5[i] = 1
    else:
        D5[i] = D5[i] + 1
        
print(D5)
```
