**Control statements** changes the flow of execution from the normal flow

They are classified into 2 types they are,
1. Break Statement
2. Continue Statement

# Break Statement
- It breaks the nearest loop when an external conditon is triggred.

```python
i = 0
while True:
    print(i)
    i = i + 1
    
    if i == 5:
        break  # here it helps us in breaking / stopping the loop
        
        # It works with the inner most condition in case of a nested loop
    
```

**Q. Find the nth odd number?**

```python
n = int(input("Enter the desired odd number you would like to find"))

i = 1
counter = 0

 
while True:
    if i % 2 == 1:             # checking if the value of i is odd
        counter = counter + 1  # every time we encounter an odd number this counter will increment by 1
    
    if counter == n:           # break condition
        print(i)
        break
        
    i = i + 1  
```

# Continue Statement
- It skips/rejects the code in current iteration of the loop and moves control to the top of the loop

```python
i = 0 

while i <= 5:
    i = i + 1
    
    if i == 3:
        continue
    print(i) 
```

**Q. Removing specific character from a string?**
