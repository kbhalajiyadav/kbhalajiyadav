# Format function

```python
# For formating the strings

a = "I am learning {}"  # Format to initialize the string formatting

b = a.format("Python") # As strings are immutable we need to store it in a variable to access it for later point of time

c = a.format("SQL")

print(b)
print(c)
```

```python
# Formatting mutiple substrings in a string

s = "{} is a {} company {}"

s1 = s.format("google","Tech",".")

print(s1)
```

```python
s = "{0} is a {1} company in {2}"

ss = "{2} is a {0} company in {2}"

s1 = s.format("Google","Tech","USA")

s2 = ss.format("Google","Tech","USA")

print(s1)
print(s2) 

# Here we can observe mutiple ways of formatting string with assigned positions as well
```

```python
# Here the format function is used with predefined labels in the string to make appropriate changes
s = "{Company_name} is a {Company_type} Company"
p = s.format(Company_type = "Tech",Company_name = "Google")

print(p)
```

```python
d = {
    "Apple":10839830,
    "Google":92294294,
    "Microsoft":282949,
    "Facebook":29794,
    "Amazon":198298,
}

for i in d:
    print("{:<10} - {:>10}".format(i,d[i]))

# Here in the above case :< (left alignment), :> (right alignment), [number] (is used for uniform spacing) is used to align 
    
# a geenric reprsentation of format function is  str.format()
```

```python
a = "The profit genrated by Apple is {:>15,}"
b = a.format(123456786)
b

# here we are using :, to insert comma.
```
