# Tuple
- Data Stored in an ordered/sequencial manner
- It is Contiguous memory allocation
- similar to list
- Immutable Data Structure (like strings) compared to list
- Operations like Del, update, append etc. cannot be performed
- Mostly used to store smaller amount of data as compared to list

```python
A = ()        # this is how we delcare an empty tuple
print(type(A))
```

```python
t = ("BY",2.3,24,"Tuple",3.7) # In tuple elements can be of any data type

print(type(t))
print(t)
```

```python
a = ("BY",2.3,24,"Tuple",3.7,["Blue","Red"],(2,3))

print(a[5][0])
print(a[6][1])

# here as stated before we can observe that tuple can conatin any data type
```

```python
a[1] = 0 

# Here we can observe that tuple are immutable (which means they can not be updated)
```

```python
a[5][0] = "Green"

# Here we can observe that mutation/ item updation has occured here, as list are mutable which is inside the tuple

print(a)
```

# Looping

```python
for i in a:
    print(i)    # it is similar to list
```

```python
l = len(a)
l
i = 0

while l > 0 :
    print(a[i])
    i = i + 1    # incrementing i to go to next element
    l = l -1     # decrementing the value of l, to iterate only 7 times
```

# Zip Function
-  It is used to connect to things
-  It takes arguments as iterable and return an iterable

```python
l1 = ["Akash", "Ameer", "Anavar", "Avrang"]
l2 = [12,12,31,22]
c = zip(l1,l2)

print(c)     # Here we can see it is been zipped which has to be converted into a list / dictionary / tuple (list(), dict(), tuple())

final_list = list(c)

print(final_list)       
```

```python
for i,j in final_list:
    print(i,j)
```

```python
l9 = [["a",3],["b",5],["c",6]] 

for i in l9:              # assign mulpitple variables simultaniously
    print(i)

for i,j in l9:
    print(i,j)
    
    
# symentricity is required in this above loops to perform them
```
