# Methods on List

```python
marks = [22,23,24,22,21,27]  # defining a list
```

**Append method**

```python
marks.append(22) # It helps to add the data to the list
```

```python
print(marks) # here we can observe that 22 has been added
```

**Insert method**

```python
marks.insert(5,31)  # insert (index, data). It is useful to add data in a particular index postion
```

```python
print(marks) # Here we can see that in index postion of 5 the value has been added.
```

```python
marks[0] = 21  # lists are mutable
```

```python
print(marks) # Here we can observe that the item in index 0 has been from 22 to 21
```

**Merge 2 lists**

```python
marks2 = [10, 22, 12, 15, 22] # defining a list
```

**Extend method**

```python
marks.extend(marks2)  # joinning 2 lists
marks
```

**Pop method**

```python
marks.pop() # by default it will pop the last element
print(marks)
```

```python
marks.pop(5) # mentioning the index 
print(marks) # here 31 has been removed
```

**Remove method**
- It removes the first occurrence of the object from the list. 

Note: If a list contains duplicate elements, it removes the first occurrence of the object from the list

```python
marks.remove(27) # It removes specified data
print(marks)     # Here we can observe the index 5 (27) has been removed
```

**count method**

```python
marks.count(15)
```

**Delete operator**

```python
del marks[3] # here it will delete the index 3 item (22)
print(marks)
```

**Generic functions**

```python
print(sum(marks)) # this are genric functions that we can use, in case of specific functions we use [dot] operator to use them
print(len(marks))

print(max(marks))
print(min(marks))

# finding the average of marks

avg = sum(marks)/ len(marks)
print("The average marks are", avg)
```

# Sorting list

**sort method**

```python
marks.sort()
print(marks) # Here we can observe the change in order of the list in ascending order
```

```python
marks.sort(reverse = True) # by default it is assending where reverse = False 
print(marks) # Here we can observe the change in order of the list in descending order
```

```python
marks.reverse() 
print(marks) # here we are reversing the overall list
```

# Silcing in List

```python
print(marks)

print(marks[:5])  # by default 1st index will be 0

print(marks[2:])  # by default the last index is the last element

print(marks[2:6]) # in order to get a specific range of index values
```

```python
# accessing every aletrnative element
print(marks)

print(marks[::2])   # here it will skip 2 elemets for the total list of items
print(marks[:10:3]) # here it will skip 2 elements for the specified range
print(marks[3::3])  # here it will skip 3 elements for the specified range
```
