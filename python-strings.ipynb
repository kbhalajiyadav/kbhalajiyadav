# **String Indexing**
- The postive indexing starts from 0, 
  where negative indexing starts from -1 (reverse indexing).
- In this way we will able to access the individual element of the string.

```python
print(a[1]) # Here h is the 2nd element
print(a[0]) # 1st element in the string

# length (len()) function can be used in order find the length of the string
print(len(a)) # here the indexing starts from 0 so it will be 8 which means (len(a) - 1)

print(a[-1]) # here we are using reverse indexing, alternative to call the last element
```

# **String Slicing**

```python
b = "Python"
```

```python
# P y t h o n
# 0 1 2 3 4 5

print(b[0:4])  # b [start : end : step]
print(b[2:])   # from 2
print(b[:5])   # from 5 backwards 

print(b[-5:-3]) # retuns in between the range

print(b[0:7:2]) # it skips every 2nd element

print(b[0:7:1]) 
```

# Inbuilt functions in string

```python
password = "1R^!,"
print(password.isalnum()) # al stands for alphabets, num for numbers

# we can observe in webistes were few states and few suggets not to use special characters 

s = "Hello"
print(s.isalpha()) # To know if all the charecters in Alphabets
```

```python
name = "Bhalaji"
print(name.isupper()) # to know if alphabets is in upper case

name = "Bhalaji"
print(name.islower()) # to know if alphabets is in lower case
```

```python
city = "Tirupati"
city = city.upper() # To change the style to upper case

print(city)

city = "Tirupati"
city = city.lower() # To change the style to lower case

print(city)
```

```python
country = "india is"
country = country.capitalize() # To capatilize the intial letter in a string

print(country)
```

```python
name = "bhaLaji K"

name = name.swapcase() # Here it is used to swap the case of the alpha in string

print(name)
```

```python
name = "bhaLaji k"
name = name.title() # It is used to display the name in proper case format
                    # It is smart enough to check space and give an proper format

print(name)
```

```python
print(name.startswith("b")) # used to know if the string starts or ends with a partciular character / letter
print(name.endswith("K"))
```

# Reverse a string

```python
a = input("Enter a sentence to reverse")

l = len(a)
r = " "

while l > 0:
    r = r + a[l-1] # As the indexing start from 0, now which consideres the last value as the beginning
    print(r)       # creating a flag to debug the code
    l = l - 1
print(r, " This is the final answer")
```
