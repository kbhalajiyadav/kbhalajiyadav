**Q. Write a function to count the number of uppercase character in a string**

```python
def upper_count(name):
    count = 0
    for i in name:
        if i.isupper():
            count = count + 1
    return count
```

```python
Name = input("Enter your name")
c = upper_count(Name)
print("number of upper case in the string is",c)
```

**Q. Write a caluclator function**

```python
def caluclator(a,b,c):
    
    '''
    Here we have 2 variables and 1 string input, to perfom the addition, subtraction, division and multiplication
    '''
    
    if c == "+":
        return a+b
    if c == "-":
        return a-b
    if c == "/":
        return a/b
    if c == "*":
        return a*b
```

# Default Argument

```python
def month_salary(weekly_hrs, weeks, pay_per_hour = 50):
    salary = weekly_hrs * weeks * pay_per_hour
    return salary
```

```python
# default argument (predefined value)
salary = month_salary(5,10)
print("Orginal Salary",salary)

# over writing predefined value
salary = month_salary(5,10, 100)
print("Hiked salary",salary)

# we can have a default value predefined, which can be over written as per the requirement
```

# Variable number of arguments

```python
def avg_grade(*marks):    # * is used to have perform with any given number of values, where here marks collects the number of values / data points.
    '''
    caluclate average marks of the student
    '''
    total, count = 0, 0 
  
    for i in marks:
        total = total + i # performing additon
        count = count + 1 # performing the count
    return total / count  # performing average
```

```python
avg = avg_grade(20,100,22)
print("The average marks of a student is", avg)
```
