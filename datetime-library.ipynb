# Time Delta Module

- It is used to find difference between datetime. It is also used to add/subract days/hours/minutes from datetime

```python
import datetime
```

**Add/Subtract**

```python
a = datetime.datetime(2020, 12, 23, 13, 14, 45)
b = a + datetime.timedelta(days = 10, minutes = 10)

# timedelta(days=0, seconds=0, microseconds=0, milliseconds=0, minutes=0, hours=0, weeks=0)

print(b)
```

```python
c = datetime.date(2020, 12, 3)
d = c - datetime.timedelta(days = 3)

print(d)
```

```python
f = b - a     # Here finding difference

print(f)
```

```python
g = c - d     # Here finding difference

print(g)
```

```python
h =  f - g     # Here finding difference

print(h)
```
