# Sets 

- Collection of unique elements
- No index based accessing
- No order in elements  (like dictionary)

      items in it are unchangable, unordered, unindexed


- Used for operations like union, insertion, difference etc

      sets can only have immutable items like integer, float, tuple, string etc
      
      They are still mutable like we can add, remove, clear the sets.

```python
s = set()      # This is how we initialize a set

print(type(s))


s = {"India", "Germany", "Japan", "Brazil"}  # here we can observe the result, which say it as set as dictionary consists of 2 values per item. 

print(type(s))

# In order to make it much clear and appropriate we use set() to intialize a set
```

```python
print(s[0])           # The we can observe that they are not subscriptable
print(s["India"])
```

# Intersection
- It is denoted by a inverted U
- Intersection will return all elements which are present in both sets

  example: 
  
      A = [2,3,4,5]
      B = [2,4,5,1]
      Here intersection of A and B are [2,4,5]
      
      If there are no intersecting elements, then it will return Null set.


```python
a = {1,2,3,4}
b = {2,4,5,6}

c = a.intersection(b)  # Method to find intersection

d  = a & b    # method to find intersection 

print(c)      # These are symmetric operations
print(d)
```

# Union

- It is denoted by U
- Union will return all the unique elements from both the sets

```python
a = {1,2,3,4}
b = {2,4,5,6}

e = a.union(b)    # Method to find union

f = a | b        # method to find union

print(e)
print(f)          # These are symmetric operations
```

# Difference
- It is denoted by minus sign (-)
- It will return a set with elements of 1 whih are not in set 2

  example:  
  
       set1 = {1,2,3,4}
       set2 = {3,4,5,6}
  
  
    set1 - set2 = {1,2} 
       
      Here it will return a set with elements of set1 which are not in set2

    set2 - set1 = {5,6}
     
      Here it will return a set with elements of set2 which are not in set1


- Here we have to observe that it is an asmmetric operation unlike UNION & INTERSECTION.

```python
a = {1,2,3,4}
b = {2,4,5,6}

g = a - b    
h = b - a

print(g, h)   # As discussed above we can observe that these are asymmetric
```

# Symmetric Difference

-  It is denoted by ^ (it is called caret)
-  set1 ^ set2 will return the set of elements which are either of the sets, but not in their intersections.

- It is an symmetric operation
- It is opposite of intersection.

```python
a = {1,2,3,4}
b = {2,4,5,6}

i = a^b                           # methods to intialize the symmetric difference

j = a.symmetric_difference(b)  

print(i)

print(j)
```

**Check wheather a string is a palangram or not** 

*Palangram is a senstense that contains every letter in English alphabet* 

*Example: The quick brown fox jumps over the lazy dog.*

```python
a = "The quick brown fox jumps over the lazy dog"
a = a.lower()

print(a)

b = set(a)

print(b)

count = 0

for i in b:
    if i.isalpha() == True:                   # checking if it the element is alpha
        count = count + 1
if count ==  26:
    print("the string is a palangram")
    
else:
    print("The string is not palangram")
```

# Sub set

set 1 is a subset of set2,
where all the elements of set1 are the elements of set2

    set1 = {1,2,3,4}
    set2 = {1,2,3,4,5,6}

    here set1 is subset of set2, where it can contain any number if subsets.

**Program to check if a set is a subset or not**


```python
setA = set(input("Enter set1"))
setB = set(input("Enter set2"))

count = 0
if len(setA) > len(setB):          # assining the set and subset
    c = setB
    d = setA
else:
    c = setA
    d = setB
    
for i in c:                         # checking the elements in setA and SetB
    for j in d:
        if i == j:
            count = count + 1
            break
            
if count == len(c):                #  checking if the set length is equal to the count
    print(c,"is a subset of",d)
else:
    print(c,"is a subset of",d)
```
