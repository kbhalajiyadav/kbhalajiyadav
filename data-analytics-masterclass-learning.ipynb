```python
#importing libraries

import pandas as pd
```

```python
a = pd.read_csv('../input/masterclass/Walmart.csv')
a
```

```python
a.dropna(how = 'all', inplace= True)

# dropna removes the null cells
# how tell olny when all the cells are null
# The changes would be temoparary (in a way to check protect any mishappens )
# inplace helps us to make the changes permanent by making the changes (trues makes the changes, by default it will be false)
```

```python
a['Country'].unique()
#here it fetches the country column from the data 
# here to find the get the unique values we will use .unique()
```

```python
a['Country'].replace({'United States' :'USA'},inplace= True)
a
```

```python
a['Country'].unique()
```

```python
# Data Imputation
# Finding mean

m = a['Sales'].mean()
m
```

```python
a['Sales'].fillna(value = m, inplace = True)
```
