<h1> Exception handling </h1>
<body>   
It is important to close our file to make that the resources get free up. 
    
Also, there might be data loss in case we do not close the file.
    
</body>

```python
try:
    f = open("../input/poetry/Kanye_West.txt")
    1/0 
    f.close()  # This will not get executed

except:
    print("There is an issue in the try block")
    
finally:
    f.close() # Here it will always get executed 
    
# The finally block is used to close resources and perform clean-up activities.

# we need such operarions in order to  is important to close our file to make that the resources get free up.
# Also, there might be data loss in case we do not close the file.
```

```python
# We can check if the file has been close or not by using the f.readline() operation
```

```python
with open("python.txt","w") as f:     # here f is the file handler/variable 
    f.write("I am learning python")
    
# after this block gets executed then the file will automatically gets closed.

# The with statement clarifies code that previously would use try, finally blocks to ensure that clean-up code is executed. 
```

```python
x = open("./python.txt")  # Here it opens in read mode by default
print(x.readline())       # using with print() is required here.
x.close()                 # Here, it gets closed
```

```python
with open("./python.txt") as x:
    print(x.readline())
    
# Here you can observe the key differenc from both the codes
```

**Ceil function**

```python
a = -4.5
b =  4.5

print("output is",math.ceil(a), "input value is -4.5")
print("output is",math.ceil(b), "input value is 4.5")
# Here it will round to the higher integer
```

**Floor function**

```python
print("output is", math.floor(a), "input value is -4.5")
print("output is", math.floor(b), "input value is 4.5")

# Here it will round to lower integer
```

```python
print("output is", math.trunc(a), "input value is -4.5")
print("output is", math.trunc(b), "input value is 4.5")

# It removes the decimal part and gives only the integer part of the given number.(truncate)
```

**Exponential & Logarithmic functions**

```python
e = 3
math.exp(3)   # e to the power of 3
```

```python
math.log(100)  # by default the base is e
```

```python
math.log(1000,10)  # now the base is 10
```

**Trignometric Functions**

```python
print(math.sin(math.pi/2))
print(math.sin(math.radians(90)))

# 1 radian = 57.2957795 degrees  (It is pi/2 = 90 degrees)
```
