# Dictionary
-  It supports unordered collection of data values
- Mutable in nature
- Every item in dictinoary has 2 parts: Key & Value

- Cannot be accessed using the index 
- No particular order i which insertion is done
- Key must be unique
- Very optimised DS when we have to retirve value using key

      selection of the data structure type depends on the use case and requirement.

```python
age = {}  # format to intilaize the dictionary

age2 = dict()

print(type(age))
print(type(age2))
```

```python
age = {
    "Rahul" : 23,
     "Joe" : 15,
    "Venkat" : 32
}

print(type(age))
```

```python
print(age)
```

# Nested Dictionary

```python
# nested dictionary 
marks = {
    "Rahul" : 23,
     "Joe" : 15,
    "Venkat" : {"section1": 12,
                "section2": 15, 
                "section3": 22}     
}

print(marks["Venkat"]["section1"]) # fetching nested dictionary items
```

```python
# nested dictionary with list

marks = {
    "Rahul" : 23,
     "Joe" : 15,
    "Venkat" : [12,13,15]
}
print(marks["Venkat"][0]) # fetching nested dictionary, list items
```

# Looping in Dictionary

```python
marks
```

```python
for i in marks:
    print(i)
    
for i in marks:         # fetching the dictionary using for loop
    print(i, marks[i])
```

# Methods in Dictionary

```python
print(age.keys())
print(age.values())
```

```python
for i in age.keys():
    print(i)
    
for i in age.values():
    print(i)
```

```python
print(age["Rahul"])       # To avoid errors and code halting, we use get function

a = marks.get("Venkat")   # for validating a item in dictinar we can use get function 
print(a)

b = age.get("rahul")      # we can also see that keys are case sensitive, as observed in prior topics
print(b)
```
