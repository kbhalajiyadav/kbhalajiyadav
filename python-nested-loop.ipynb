**Q.Print series of numbers in a loop**

```python
n1 = int(input("enter how many times would you like to repeat the loop"))
n2 = int(input("enter how many series of numbers would you like to print"))

print("The total number of iterations that you have reuquested is", n1*n2)

i = 1
while i <= n1:
    j = 1
    while j <= n2:
        print(j,end = " ")
        j = j+1
    i = i + 1  
    print("\n")  #Prints in next line
```

```python
n1 = int(input("enter how many times would you like to repeat the loop"))
n2 = int(input("enter how many series of numbers would you like to print"))

print("The total number of iterations that you have reuquested is", n1*n2)

for i in range(n1,n2):
    j = 1
    while j <= n2:
        print(j,end = " ")
        j = j+1
    i = i + 1  
    print("\n")  #Prints in next line
```

**Q. Print ringht angle triangle pattern**

```python
n1 = int(input("enter how many times would you like to repeat the loop"))
s  = input("enter the symbol would you like to print")

i = 1
while i <= n1:
    j = 1
    while j <= i:
        print(s, end = " ")
        j = j + 1
    print("\n")
    i = i + 1
```

```python
n1 = int(input("enter how many times would you like to repeat the loop"))
s  = input("enter the symbol would you like to print")

for i in range(n1+1):
    j = 1
    while j <= i:
        print(s, end = " ")
        j = j + 1
    print("\n")
    i = i + 1
```

**Q. Print prime numbers for a given range**

```python
n1 = int(input("Enter the lower limit number"))
n2 = int(input("Enter the upper limit number"))

if n1 == 1:
    n1 = n1 +1
    
i = n1
while i <= n2:                 # range of number which we have to find weather they are prime or not
    j = n1
    while j <= i // 2 + 1:     # maximun multiple we will check
                               # here we gives the range to iterate (be i // 2 + 1) the loop 
        if i == 2:
            print(i, end = ", ")
        if i % j == 0:
            break              # here it passes over to next if, if it is not satisfied
        if j == i // 2 + 1:    # here it checks if the it have runed i // 2 + 1 times, then print
            print(i, end = ", ")
        j = j + 1
    i = i + 1
```
